```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

# Data Preparation 

```python
data = pd.read_csv('directory.csv',encoding='ISO-8859-1')
data.head()
```

```python
data.shape
```

```python
# check 
chk1= data.duplicated(subset=['Store_Number']).sum()
chk1
```

```python
data.drop_duplicates(subset=['Store_Number'],keep='first', inplace=True)
```

```python
data.isnull().sum()
```

```python
# Remove missing values 
data1= data.dropna(subset=['City'])
data1.isnull().sum()
```

```python
data1.shape
```

# Top 10 Countries 

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111)
ax.set(title = "Top 10 Countries with Most Number of Starbucks Stores")
data1.Country.value_counts().head(10).plot(kind="bar", color = "green")

p=data.Country.value_counts()/len(data.Country)
pcts = (data1.Country.value_counts(normalize=True) * 100).apply(lambda x: '{:.2f}%'.format(x))

for i, p in enumerate(ax.patches):
    ax.annotate(pcts[i], (p.get_x(), p.get_height() * 1.01))
       
plt.show()
```

# Top 10 Cities with most number of Starbucks Stores

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111)
ax.set(title = "Top 10 Cities with most number of Starbucks Stores")
data1.City.value_counts().head(10).plot(kind="bar",color = "green")

p=data.City.value_counts()/len(data.City)
pcts = (data1.City.value_counts(normalize=True) * 100).apply(lambda x: '{:.2f}%'.format(x))

for i, p in enumerate(ax.patches):
    ax.annotate(pcts[i], (p.get_x(), p.get_height() * 1.01))

plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111)
usa_states = data1[data1['Country'] == 'US']
ax.set(title="What are the Top 10 States in USA with most number of stores?")
usa_states['State/Province'].value_counts().head(10).plot(kind="bar",color = "green")

p=usa_states['State/Province'].value_counts()/len(data1.City)
pcts = (usa_states['State/Province'].value_counts(normalize=True) * 100).apply(lambda x: '{:.2f}%'.format(x))
pcts = (usa_states.City.value_counts(normalize=True) * 100).apply(lambda x: '{:.2f}%'.format(x))

for i, p in enumerate(ax.patches):
    ax.annotate(pcts[i], (p.get_x(), p.get_height() * 1.01))

plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111)
usa_states = data1[data1['Country'] == 'US']
ax.set(title="What are the Top 10 City with most number of stores?")
usa_states.City.value_counts().head(10).plot(kind="bar",color = "green")

p=usa_states.City.value_counts()/len(data1.City)
pcts = (usa_states.City.value_counts(normalize=True) * 100).apply(lambda x: '{:.2f}%'.format(x))

for i, p in enumerate(ax.patches):
    ax.annotate(pcts[i], (p.get_x(), p.get_height() * 1.01))    

plt.show()         
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
plt.figure(figsize=(10, 10))
m = Basemap(projection='ortho', resolution=None, lat_0=50, lon_0=-100)
m.bluemarble(scale=0.5);

x, y = m(list(data1["Longitude"].astype(float)), list(data1["Latitude"].astype(float)))
m.plot(x, y, 'bo', markersize = 5, alpha = 0.6, color = "green")

plt.title('Starbucks Stores Across the World')
plt.show()
```

```python
plt.figure(figsize=(15,20))
m = Basemap(projection = 'mill', llcrnrlat = -80, urcrnrlat = 80, llcrnrlon = -180, urcrnrlon = 180, resolution = 'h')
m.drawcoastlines()

m.drawmapboundary(fill_color='white')

x, y = m(list(data1["Longitude"].astype(float)), list(data1["Latitude"].astype(float)))
m.plot(x, y, 'bo', markersize = 5, alpha = 0.6, color = "green")

plt.title('Starbucks Stores Across the World')
plt.show()
```
